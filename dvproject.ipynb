```python
plt.figure(figsize=(5,5))
sns.countplot(x='Generation',data=df)
plt.title('Distribution of generation')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Type 1',data=df)
plt.title('Distribution of Type 1')
plt.show()
```

```python
Labels=df['Attack'].head(5)
color=['#ffccff','#ffb3ff','#ff99ff','#e600e6','#990099']
explode=[0,0.1,0,0,0]
plt.pie(Labels,autopct='%1.1f%%',colors=color,startangle=90,explode=explode)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Type 1',y='Total',data=df)
plt.title("distribution of Type1")
plt.xlabel('Type 1')
plt.ylabel('Total')
plt.show()
```

```python
df.corr()
```

```python
correlation_matrix=df.corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt=".2f")
plt.title("correlation heatmap")
plt.show
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Type 1',y='Speed',hue='Generation',data=df,ci=None)
plt.title("distribution of Type1")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Generation',y='Attack',data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.jointplot(x='Total',y='Speed',data=df,kind='scatter')
plt.title("sctter plot of total vs speed")
plt.xlabel('total')
plt.ylabel('speed')
plt.show()
```
